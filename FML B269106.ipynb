```python
import numpy as np
import pandas as pd
import hdf5storage
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import class_weight
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, Flatten, Input, Dropout, Bidirectional
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
import warnings
import csv
import os

warnings.filterwarnings("ignore")

# Set Seeds for Reproducibility
np.random.seed(23)
tf.random.set_seed(23)
random.seed(23)

# Dictionary to store final predictions (from the last fold) for plotting
final_preds = {}
final_actual = None
# For confusion matrix of LSTM
final_lstm_class_preds = None
final_lstm_y_true = None


# Feature Engineering & Selection Functions

def load_lob_data(path):
    """
    Load LOB data from a MATLAB .mat file and return as a DataFrame.
    Exception handling is incorporated to catch file errors.
    """
    try:
        mat_data = hdf5storage.loadmat(path)
        df = pd.DataFrame.from_dict(mat_data['LOB'])
        return df
    except Exception as e:
        print(f"Error loading file: {e}")
        raise


def handle_missing(df):
    """
    Fill missing values using forward/backward fill and drop any remaining NaNs.
    """
    try:
        for col in df.columns:
            df[col] = df[col].fillna(method='ffill').fillna(method='bfill')
        return df.dropna()
    except Exception as e:
        print(f"Error handling missing data: {e}")
        raise


def add_engineered_features(df):
    """
    Add additional features:
      - For each LOB level, compute AskDelta, VolumeSum, and PriceDiff.
      - For Mid_Price: add lag features, rolling averages, and rolling volatility.
    Enhanced documentation:
      - AskDelta: Difference between AskPrice and BidPrice for each level.
      - VolumeSum: Sum of AskVolume and BidVolume for each level.
      - PriceDiff: Redundant check (same as AskDelta) for this example.
      - MidPrice features: Lag features help capture temporal dependencies;
        rolling averages and std deviations help capture trend and volatility.
    """
    try:
        for i in range(1, 11):
            df[f'AskDelta{i}'] = df[f'AskPrice{i}'] - df[f'BidPrice{i}']
            df[f'VolumeSum{i}'] = df[f'AskVolume{i}'] + df[f'BidVolume{i}']
            df[f'PriceDiff{i}'] = df[f'AskPrice{i}'] - df[f'BidPrice{i}']
        # Mid_Price lag features
        df['MidPrice_Lag1'] = df['Mid_Price'].shift(1)
        df['MidPrice_Lag2'] = df['Mid_Price'].shift(2)
        # Rolling averages for Mid_Price
        df['MidPrice_RollingAvg_5'] = df['Mid_Price'].rolling(window=5).mean()
        df['MidPrice_RollingAvg_10'] = df['Mid_Price'].rolling(window=10).mean()
        # Rolling volatility (std) for Mid_Price
        df['MidPrice_RollingStd_5'] = df['Mid_Price'].rolling(window=5).std()
        df['MidPrice_RollingStd_10'] = df['Mid_Price'].rolling(window=10).std()
        df.dropna(inplace=True)
        return df
    except Exception as e:
        print(f"Error in feature engineering: {e}")
        raise


def select_features(df, target_column='Mid_Price', threshold=0.01):
    """
    Optionally select features that have an absolute Pearson correlation with the target above a threshold.
    Returns a DataFrame with the selected features.
    """
    try:
        corr = df.corr()[target_column].abs()
        selected = corr[corr > threshold].index.tolist()
        return df[selected]
    except Exception as e:
        print(f"Error selecting features: {e}")
        raise


def create_sliding_window(X, y_reg, y_cls, look_back):
    """
    Create 3D sliding windows for sequential models.
    Returns arrays of shape [samples, look_back, features] and corresponding targets.
    """
    Xs, yr, yc = [], [], []
    for i in range(len(X) - look_back):
        Xs.append(X[i:i + look_back])
        yr.append(y_reg[i + look_back])
        yc.append(y_cls[i + look_back])
    return np.array(Xs), np.array(yr), np.array(yc)


def plot_predictions(y_true, y_pred, title):
    """
    Plot actual vs. predicted regression values.
    """
    plt.figure(figsize=(12, 5))
    plt.plot(y_true[:1000], label='Actual', linewidth=2)
    plt.plot(y_pred[:1000], label='Predicted', alpha=0.75)
    plt.title(title)
    plt.xlabel("Samples")
    plt.ylabel("Mid Price")
    plt.legend()
    plt.tight_layout()
    plt.show()


def print_classification_metrics(y_true, y_pred, binary_classification=True):
    """
    Print classification metrics.
    For binary classification, thresholds predictions at 0.5.
    """
    if binary_classification:
        y_pred_cls = (y_pred > 0.5).astype(int).flatten()
    else:
        y_pred_cls = np.argmax(y_pred, axis=1)
    y_true_cls = y_true.flatten()
    acc = accuracy_score(y_true_cls, y_pred_cls)
    print("Classification Accuracy: {:.2f}%".format(acc * 100))
    print("Classification Report:")
    print(classification_report(y_true_cls, y_pred_cls))
    print("Confusion Matrix:")
    print(confusion_matrix(y_true_cls, y_pred_cls))



# Model Building Function

def build_dual_output_model(input_shape, model_type="mlp", depth=1, learning_rate=0.001, binary_classification=True):
    """
    Build a dual-output model that predicts both a regression target (mid-price)
    and a classification target.
    
    For binary classification (binary_classification=True):
      - Classification branch: 1 output with sigmoid activation and binary cross-entropy loss.
    
    For MLP:
      - depth=1: Shallow (1 hidden layer)
      - depth=2: Deep (2 hidden layers)
      - depth>=3: Very Deep (3 hidden layers)
      (L2 regularization is applied in Dense layers.)
    
    For sequential models, input_shape should be (look_back, features).
    """
    inputs = Input(shape=input_shape)

    if model_type == "mlp":
        # Use L2 regularization (0.001) with Dropout to prevent overfitting
        if depth == 1:
            x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(inputs)
            x = Dropout(0.3)(x)
        elif depth == 2:
            x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(inputs)
            x = Dropout(0.3)(x)
            x = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
            x = Dropout(0.3)(x)
        elif depth >= 3:
            x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(inputs)
            x = Dropout(0.3)(x)
            x = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
            x = Dropout(0.3)(x)
            x = Dense(64, activation='relu', kernel_regularizer=l2(0.001))(x)
            x = Dropout(0.3)(x)

    elif model_type == "lstm":
        x = Bidirectional(LSTM(128, return_sequences=True, recurrent_dropout=0.2))(inputs)
        x = Bidirectional(LSTM(64, return_sequences=False, recurrent_dropout=0.2))(x)

    elif model_type == "cnn":
        x = Conv1D(64, 3, activation='relu', padding='same')(inputs)
        x = MaxPooling1D(pool_size=2)(x)
        x = Conv1D(32, 3, activation='relu', padding='same')(x)
        x = Flatten()(x)

    elif model_type == "cnn_lstm":
        x = Conv1D(64, 3, activation='relu', padding='same')(inputs)
        x = MaxPooling1D(pool_size=2)(x)
        x = LSTM(64, recurrent_dropout=0.2)(x)

    # Regression output: Predicts continuous mid-price value
    out_reg = Dense(1, name='regression')(x)
    # Classification output: Predicts binary (or multi-class) movement direction
    if binary_classification:
        out_cls = Dense(1, activation='sigmoid', name='classification')(x)
    else:
        out_cls = Dense(3, activation='softmax', name='classification')(x)

    model = Model(inputs=inputs, outputs={'regression': out_reg, 'classification': out_cls})
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

    if binary_classification:
        losses = {'regression': 'mse', 'classification': 'binary_crossentropy'}
        loss_weights = {'regression': 0.3, 'classification': 0.7}  # Emphasize classification performance
    else:
        losses = {'regression': 'mse', 'classification': 'sparse_categorical_crossentropy'}
        loss_weights = {'regression': 0.7, 'classification': 0.3}

    model.compile(
        loss=losses,
        metrics={'regression': ['mse'], 'classification': ['accuracy']},
        loss_weights=loss_weights,
        optimizer=optimizer
    )
    return model



# Main Script

# File path updated for the MATLAB file.
file_path = "C:/Users/Tilak Heble/OneDrive/Desktop/FML Assignment/S092215-v50-AMZN_OCT2_states.mat"
try:
    df = load_lob_data(file_path)
except Exception as e:
    print("Exiting due to file load error.")
    exit(1)

# Set column names: "Time", "Mid_Price", "Spread", then for each level: AskPrice{i}, AskVolume{i}, BidPrice{i}, BidVolume{i}
cols = ["Time", "Mid_Price", "Spread"] + [f"{s}{i}" for i in range(1, 11) for s in ("AskPrice", "AskVolume", "BidPrice", "BidVolume")]
df.columns = cols

try:
    # Preprocess data with exception handling
    df = handle_missing(df)
    # Use a subset for quicker runs; adjust indices as needed
    df = df.iloc[300000:450000, :]
    df = add_engineered_features(df)
    # Optionally, perform feature selection
    df = select_features(df, target_column='Mid_Price', threshold=0.01)
except Exception as e:
    print("Error during preprocessing:", e)
    exit(1)

# Redefine Classification Target for Binary Classification
# Define binary target: 1 if next percentage change in Mid_Price is positive, else 0.
df['Mid_Price_Change'] = df['Mid_Price'].pct_change().shift(-1)
df['Mid_Price_Class'] = df['Mid_Price_Change'].apply(lambda x: 1 if x > 0 else 0)
df.dropna(inplace=True)

# Define features and targets
feature_cols = [col for col in df.columns if col not in ['Time', 'Mid_Price_Change', 'Mid_Price_Class']]
X_raw = df[feature_cols]
# Regression target: log1p-transformed Mid_Price (reverse later with expm1)
y_reg_raw = np.log1p(df["Mid_Price"].values.reshape(-1, 1))
y_cls_raw = df["Mid_Price_Class"].values.reshape(-1, 1)



# Hyperparameter Tuning Setup

# For MLP, tune three variants: shallow (depth=1), deep (depth=2), and very deep (depth=3)
mlp_depths = [1, 2, 3]
batch_sizes = [256]       # Single option for quick tuning
epochs_list = [10]        # Fewer epochs for quick runs
learning_rates = [0.001]  # Single learning rate option

# Use 5-fold TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)
fold = 0
early_stop = EarlyStopping(patience=2, restore_best_weights=True)

# Dictionaries to store results for each model type
results = {"MLP_shallow": [], "MLP_deep": [], "MLP_very_deep": [], "LSTM": [], "CNN": [], "CNN-LSTM": []}
# List to store ensemble results (RMSE, MAE) per fold
ensemble_results = []

# For storing best sequential models for ensemble later.
ensemble_models = {}

# Loop through the time series folds
for train_idx, test_idx in tscv.split(X_raw):
    fold += 1
    print(f"\n========== Fold {fold} ==========")

    try:
        # Split data
        X_train_raw = X_raw.iloc[train_idx]
        X_test_raw = X_raw.iloc[test_idx]
        y_train_reg_raw = y_reg_raw[train_idx]
        y_test_reg_raw = y_reg_raw[test_idx]
        y_train_cls_raw = y_cls_raw[train_idx]
        y_test_cls_raw = y_cls_raw[test_idx]

        # Scale features and regression target
        X_scaler = MinMaxScaler()
        y_scaler = MinMaxScaler()
        X_train = X_scaler.fit_transform(X_train_raw)
        X_test = X_scaler.transform(X_test_raw)
        y_train_reg = y_scaler.fit_transform(y_train_reg_raw)
        y_test_reg = y_scaler.transform(y_test_reg_raw)

        # Compute class weights for binary classification
        weights = class_weight.compute_class_weight(
            class_weight='balanced',
            classes=np.unique(y_train_cls_raw.flatten()),
            y=y_train_cls_raw.flatten()
        )
        class_weights_dict = dict(enumerate(weights))
        sample_weights_cls = np.array([class_weights_dict.get(int(label), 1.0) for label in y_train_cls_raw.flatten()])

        input_dim = X_train.shape[1]
    except Exception as e:
        print(f"Error during data split/scaling in fold {fold}: {e}")
        continue

    # ---- MLP Variants ----
    for depth in mlp_depths:
        if depth == 1:
            model_label = "MLP_shallow"
        elif depth == 2:
            model_label = "MLP_deep"
        elif depth >= 3:
            model_label = "MLP_very_deep"

        print(f"\n--- Tuning {model_label} Model ---")
        best_rmse = np.inf
        best_params = None
        for batch_size in batch_sizes:
            for epochs in epochs_list:
                for lr in learning_rates:
                    try:
                        model = build_dual_output_model(input_shape=(input_dim,), model_type="mlp", depth=depth, learning_rate=lr, binary_classification=True)
                        model.fit(
                            X_train,
                            {'regression': y_train_reg, 'classification': y_train_cls_raw},
                            sample_weight={
                                'regression': np.ones(len(sample_weights_cls), dtype=np.float32),
                                'classification': sample_weights_cls.astype(np.float32)
                            },
                            epochs=epochs,
                            batch_size=batch_size,
                            verbose=0,
                            callbacks=[early_stop]
                        )
                        y_pred = model.predict(X_test)
                        # Reverse log1p transformation for regression predictions and targets
                        y_pred_reg = np.expm1(y_scaler.inverse_transform(y_pred['regression']))
                        y_test_reg_inv = np.expm1(y_scaler.inverse_transform(y_test_reg))
                        # Calculate RMSE and MAE only
                        rmse_val = np.sqrt(np.mean((y_test_reg_inv - y_pred_reg) ** 2))
                        mae_val = mean_absolute_error(y_test_reg_inv, y_pred_reg)
                        if rmse_val < best_rmse:
                            best_rmse = rmse_val
                            best_params = (depth, batch_size, epochs, lr, mae_val)
                    except Exception as e:
                        print(f"Error during MLP tuning in fold {fold}: {e}")
                        continue
        if best_params is None:
            continue

        print(f"Best {model_label} params: Depth={best_params[0]}, Batch Size={best_params[1]}, Epochs={best_params[2]}, LR={best_params[3]}")
        print(f"Best {model_label} Metrics -- RMSE: {best_rmse:.6f}, MAE: {best_params[4]:.6f}")
        results[model_label].append((best_rmse, best_params[4]))

        try:
            best_model = build_dual_output_model(input_shape=(input_dim,), model_type="mlp", depth=best_params[0], learning_rate=best_params[3], binary_classification=True)
            best_model.fit(
                X_train,
                {'regression': y_train_reg, 'classification': y_train_cls_raw},
                sample_weight={
                    'regression': np.ones(len(sample_weights_cls), dtype=np.float32),
                    'classification': sample_weights_cls.astype(np.float32)
                },
                epochs=best_params[2],
                batch_size=best_params[1],
                verbose=0,
                callbacks=[early_stop]
            )
            y_pred = best_model.predict(X_test)
            y_pred_reg = np.expm1(y_scaler.inverse_transform(y_pred['regression']))
            y_test_reg_inv = np.expm1(y_scaler.inverse_transform(y_test_reg))
            plot_predictions(y_test_reg_inv.flatten(), y_pred_reg.flatten(), f"Best {model_label} - Fold {fold}")
            print(f"Classification Metrics for Best {model_label} - Fold {fold}:")
            print_classification_metrics(y_test_cls_raw, y_pred['classification'], binary_classification=True)
            if fold == tscv.get_n_splits():
                final_preds[model_label] = y_pred_reg.flatten()
                final_actual = y_test_reg_inv.flatten()
        except Exception as e:
            print(f"Error during final training for {model_label} in fold {fold}: {e}")

    # ---- Sequential Models: Create Sliding Windows ----
    try:
        look_back = 10
        input_seq_shape = (look_back, input_dim)
        X_train_seq, y_train_r, y_train_c = create_sliding_window(X_train, y_train_reg, y_train_cls_raw, look_back)
        X_test_seq, y_test_r, y_test_c = create_sliding_window(X_test, y_test_reg, y_test_cls_raw, look_back)
        sw_seq = sample_weights_cls[look_back:]
    except Exception as e:
        print(f"Error creating sliding windows in fold {fold}: {e}")
        continue

    # ---- LSTM Model ----
    print("\n--- Tuning LSTM Model ---")
    best_lstm_rmse = np.inf
    best_lstm_params = None
    for batch_size in batch_sizes:
        for epochs in epochs_list:
            for lr in learning_rates:
                try:
                    model = build_dual_output_model(input_shape=input_seq_shape, model_type="lstm", learning_rate=lr, binary_classification=True)
                    model.fit(
                        X_train_seq,
                        {'regression': y_train_r, 'classification': y_train_c},
                        sample_weight={
                            'regression': np.ones(len(sw_seq), dtype=np.float32),
                            'classification': sw_seq.astype(np.float32)
                        },
                        epochs=epochs,
                        batch_size=batch_size,
                        verbose=0,
                        callbacks=[early_stop]
                    )
                    y_pred = model.predict(X_test_seq)
                    y_pred_reg = np.expm1(y_scaler.inverse_transform(y_pred['regression']))
                    y_test_r_inv = np.expm1(y_scaler.inverse_transform(y_test_r))
                    rmse_val = np.sqrt(np.mean((y_test_r_inv - y_pred_reg) ** 2))
                    mae_val = mean_absolute_error(y_test_r_inv, y_pred_reg)
                    if rmse_val < best_lstm_rmse:
                        best_lstm_rmse = rmse_val
                        best_lstm_params = (batch_size, epochs, lr, mae_val)
                except Exception as e:
                    print(f"Error during LSTM tuning in fold {fold}: {e}")
                    continue
    if best_lstm_params is not None:
        print(f"Best LSTM params: Batch Size={best_lstm_params[0]}, Epochs={best_lstm_params[1]}, LR={best_lstm_params[2]}")
        print(f"Best LSTM Metrics -- RMSE: {best_lstm_rmse:.6f}, MAE: {best_lstm_params[3]:.6f}")
        results["LSTM"].append((best_lstm_rmse, best_lstm_params[3]))
        try:
            best_lstm_model = build_dual_output_model(input_shape=input_seq_shape, model_type="lstm", learning_rate=best_lstm_params[2], binary_classification=True)
            best_lstm_model.fit(
                X_train_seq,
                {'regression': y_train_r, 'classification': y_train_c},
                sample_weight={
                    'regression': np.ones(len(sw_seq), dtype=np.float32),
                    'classification': sw_seq.astype(np.float32)
                },
                epochs=best_lstm_params[1],
                batch_size=best_lstm_params[0],
                verbose=0,
                callbacks=[early_stop]
            )
            y_pred = best_lstm_model.predict(X_test_seq)
            y_pred_reg = np.expm1(y_scaler.inverse_transform(y_pred['regression']))
            y_test_r_inv = np.expm1(y_scaler.inverse_transform(y_test_r))
            plot_predictions(y_test_r_inv.flatten(), y_pred_reg.flatten(), f"Best LSTM - Fold {fold}")
            print(f"Classification Metrics for Best LSTM - Fold {fold}:")
            print_classification_metrics(y_test_c, y_pred['classification'], binary_classification=True)
            if fold == tscv.get_n_splits():
                final_preds["LSTM"] = y_pred_reg.flatten()
                final_actual = y_test_r_inv.flatten()
                final_lstm_class_preds = (y_pred['classification'] > 0.5).astype(int).flatten()
                final_lstm_y_true = y_test_c.flatten()
        except Exception as e:
            print(f"Error during final training for LSTM in fold {fold}: {e}")

    # ---- CNN Model ----
    print("\n--- Tuning CNN Model ---")
    best_cnn_rmse = np.inf
    best_cnn_params = None
    for batch_size in batch_sizes:
        for epochs in epochs_list:
            for lr in learning_rates:
                try:
                    model = build_dual_output_model(input_shape=input_seq_shape, model_type="cnn", learning_rate=lr, binary_classification=True)
                    model.fit(
                        X_train_seq,
                        {'regression': y_train_r, 'classification': y_train_c},
                        sample_weight={
                            'regression': np.ones(len(sw_seq), dtype=np.float32),
                            'classification': sw_seq.astype(np.float32)
                        },
                        epochs=epochs,
                        batch_size=batch_size,
                        verbose=0,
                        callbacks=[early_stop]
                    )
                    y_pred = model.predict(X_test_seq)
                    y_pred_reg = np.expm1(y_scaler.inverse_transform(y_pred['regression']))
                    y_test_r_inv = np.expm1(y_scaler.inverse_transform(y_test_r))
                    rmse_val = np.sqrt(np.mean((y_test_r_inv - y_pred_reg) ** 2))
                    mae_val = mean_absolute_error(y_test_r_inv, y_pred_reg)
                    if rmse_val < best_cnn_rmse:
                        best_cnn_rmse = rmse_val
                        best_cnn_params = (batch_size, epochs, lr, mae_val)
                except Exception as e:
                    print(f"Error during CNN tuning in fold {fold}: {e}")
                    continue
    if best_cnn_params is not None:
        print(f"Best CNN params: Batch Size={best_cnn_params[0]}, Epochs={best_cnn_params[1]}, LR={best_cnn_params[2]}")
        print(f"Best CNN Metrics -- RMSE: {best_cnn_rmse:.6f}, MAE: {best_cnn_params[3]:.6f}")
        results["CNN"].append((best_cnn_rmse, best_cnn_params[3]))
        try:
            best_cnn_model = build_dual_output_model(input_shape=input_seq_shape, model_type="cnn", learning_rate=best_cnn_params[2], binary_classification=True)
            best_cnn_model.fit(
                X_train_seq,
                {'regression': y_train_r, 'classification': y_train_c},
                sample_weight={
                    'regression': np.ones(len(sw_seq), dtype=np.float32),
                    'classification': sw_seq.astype(np.float32)
                },
                epochs=best_cnn_params[1],
                batch_size=best_cnn_params[0],
                verbose=0,
                callbacks=[early_stop]
            )
            y_pred = best_cnn_model.predict(X_test_seq)
            y_pred_reg = np.expm1(y_scaler.inverse_transform(y_pred['regression']))
            y_test_r_inv = np.expm1(y_scaler.inverse_transform(y_test_r))
            plot_predictions(y_test_r_inv.flatten(), y_pred_reg.flatten(), f"Best CNN - Fold {fold}")
            print(f"Classification Metrics for Best CNN - Fold {fold}:")
            print_classification_metrics(y_test_c, y_pred['classification'], binary_classification=True)
            if fold == tscv.get_n_splits():
                final_preds["CNN"] = y_pred_reg.flatten()
                final_actual = y_test_r_inv.flatten()
        except Exception as e:
            print(f"Error during final training for CNN in fold {fold}: {e}")

    # ---- CNN-LSTM Model ----
    print("\n--- Tuning CNN-LSTM Model ---")
    best_cnn_lstm_rmse = np.inf
    best_cnn_lstm_params = None
    for batch_size in batch_sizes:
        for epochs in epochs_list:
            for lr in learning_rates:
                try:
                    model = build_dual_output_model(input_shape=input_seq_shape, model_type="cnn_lstm", learning_rate=lr, binary_classification=True)
                    model.fit(
                        X_train_seq,
                        {'regression': y_train_r, 'classification': y_train_c},
                        sample_weight={
                            'regression': np.ones(len(sw_seq), dtype=np.float32),
                            'classification': sw_seq.astype(np.float32)
                        },
                        epochs=epochs,
                        batch_size=batch_size,
                        verbose=0,
                        callbacks=[early_stop]
                    )
                    y_pred = model.predict(X_test_seq)
                    y_pred_reg = np.expm1(y_scaler.inverse_transform(y_pred['regression']))
                    y_test_r_inv = np.expm1(y_scaler.inverse_transform(y_test_r))
                    rmse_val = np.sqrt(np.mean((y_test_r_inv - y_pred_reg) ** 2))
                    mae_val = mean_absolute_error(y_test_r_inv, y_pred_reg)
                    if rmse_val < best_cnn_lstm_rmse:
                        best_cnn_lstm_rmse = rmse_val
                        best_cnn_lstm_params = (batch_size, epochs, lr, mae_val)
                except Exception as e:
                    print(f"Error during CNN-LSTM tuning in fold {fold}: {e}")
                    continue
    if best_cnn_lstm_params is not None:
        print(f"Best CNN-LSTM params: Batch Size={best_cnn_lstm_params[0]}, Epochs={best_cnn_lstm_params[1]}, LR={best_cnn_lstm_params[2]}")
        print(f"Best CNN-LSTM Metrics -- RMSE: {best_cnn_lstm_rmse:.6f}, MAE: {best_cnn_lstm_params[3]:.6f}")
        results["CNN-LSTM"].append((best_cnn_lstm_rmse, best_cnn_lstm_params[3]))
        try:
            best_cnn_lstm_model = build_dual_output_model(input_shape=input_seq_shape, model_type="cnn_lstm", learning_rate=best_cnn_lstm_params[2], binary_classification=True)
            best_cnn_lstm_model.fit(
                X_train_seq,
                {'regression': y_train_r, 'classification': y_train_c},
                sample_weight={
                    'regression': np.ones(len(sw_seq), dtype=np.float32),
                    'classification': sw_seq.astype(np.float32)
                },
                epochs=best_cnn_lstm_params[1],
                batch_size=best_cnn_lstm_params[0],
                verbose=0,
                callbacks=[early_stop]
            )
            y_pred = best_cnn_lstm_model.predict(X_test_seq)
            y_pred_reg = np.expm1(y_scaler.inverse_transform(y_pred['regression']))
            y_test_r_inv = np.expm1(y_scaler.inverse_transform(y_test_r))
            plot_predictions(y_test_r_inv.flatten(), y_pred_reg.flatten(), f"Best CNN-LSTM - Fold {fold}")
            print(f"Classification Metrics for Best CNN-LSTM - Fold {fold}:")
            print_classification_metrics(y_test_c, y_pred['classification'], binary_classification=True)
            if fold == tscv.get_n_splits():
                final_preds["CNN-LSTM"] = y_pred_reg.flatten()
                final_actual = y_test_r_inv.flatten()
        except Exception as e:
            print(f"Error during final training for CNN-LSTM in fold {fold}: {e}")

    # ---- Ensemble of LSTM and CNN-LSTM ----
    print("\n--- Ensemble (LSTM + CNN-LSTM) ---")
    try:
        # Obtain predictions from the best LSTM and CNN-LSTM models
        y_pred_lstm = best_lstm_model.predict(X_test_seq)
        y_pred_cnn_lstm = best_cnn_lstm_model.predict(X_test_seq)
        # Ensemble regression: average predictions
        y_pred_reg_ensemble = (y_pred_lstm['regression'] + y_pred_cnn_lstm['regression']) / 2
        y_pred_reg_ensemble = np.expm1(y_scaler.inverse_transform(y_pred_reg_ensemble))
        y_test_reg_inv = np.expm1(y_scaler.inverse_transform(y_test_r))
        ensemble_rmse = np.sqrt(np.mean((y_test_reg_inv - y_pred_reg_ensemble) ** 2))
        ensemble_mae = mean_absolute_error(y_test_reg_inv, y_pred_reg_ensemble)
        print("Ensemble Regression Metrics (LSTM & CNN-LSTM): RMSE: {:.2f}, MAE: {:.2f}".format(ensemble_rmse, ensemble_mae))
        # Ensemble classification: average the predicted probabilities and threshold at 0.5
        y_pred_cls_ensemble = (y_pred_lstm['classification'] + y_pred_cnn_lstm['classification']) / 2
        print("Ensemble Classification Metrics:")
        print_classification_metrics(y_test_c, y_pred_cls_ensemble, binary_classification=True)
        ensemble_results.append((ensemble_rmse, ensemble_mae))
        if fold == tscv.get_n_splits():
            final_preds["Ensemble"] = y_pred_reg_ensemble.flatten()
            final_actual = y_test_reg_inv.flatten()
    except Exception as e:
        print(f"Error during ensemble predictions in fold {fold}: {e}")


# Save Overall Results to CSV

try:
    results_list = []
    for model_type, metrics in results.items():
        for i, (rmse_val, mae_val) in enumerate(metrics, start=1):
            results_list.append({
                "Model": model_type,
                "Fold": i,
                "RMSE": rmse_val,
                "MAE": mae_val
            })
    # Add ensemble results to the CSV
    for i, (rmse_val, mae_val) in enumerate(ensemble_results, start=1):
        results_list.append({
            "Model": "Ensemble",
            "Fold": i,
            "RMSE": rmse_val,
            "MAE": mae_val
        })
    results_df = pd.DataFrame(results_list)
    results_csv_path = "results_summary.csv"
    results_df.to_csv(results_csv_path, index=False)
    print(f"\nOverall results saved to {os.path.abspath(results_csv_path)}")
except Exception as e:
    print(f"Error saving results to CSV: {e}")

# Print overall results per model type across folds
print("\n=== Overall Results (per fold) ===")
for model_type, metrics in results.items():
    print(f"{model_type}:")
    for i, (rmse_val, mae_val) in enumerate(metrics, start=1):
        print(f"  Fold {i} - RMSE: {rmse_val:.2f}, MAE: {mae_val:.2f}")

# Print overall ensemble results across folds
if ensemble_results:
    print("\n=== Overall Ensemble Results (per fold) ===")
    for i, (rmse_val, mae_val) in enumerate(ensemble_results, start=1):
        print(f"  Fold {i} - RMSE: {rmse_val:.2f}, MAE: {mae_val:.2f}")


import seaborn as sns

# Define which three models you want to show in Figure 1
best_3 = ["LSTM", "CNN-LSTM", "Ensemble"] 

# Figure 1: Actual vs. Predicted Mid-Price Plot for the Best 3 Models
plt.figure(figsize=(14, 8))
if final_actual is not None:
    plt.plot(final_actual[:1000], label="Actual", linewidth=2)

for key in best_3:
    if key in final_preds:
        plt.plot(final_preds[key][:1000], label=f"Predicted {key}", alpha=0.75)

plt.title("Figure 1: Actual vs. Predicted Mid-Price for Best 3 Models")
plt.xlabel("Samples")
plt.ylabel("Mid-Price")
plt.legend()
plt.tight_layout()
plt.show()

# Figure 2: Confusion Matrix Heatmap for the Best LSTM Model
if final_lstm_y_true is not None and final_lstm_class_preds is not None:
    cm = confusion_matrix(final_lstm_y_true, final_lstm_class_preds)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title("Figure 2: Confusion Matrix Heatmap for Best LSTM Model")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()
    plt.show()

# Figure 3: Bar Chart Comparing Average RMSE and MAE Across Different Models
avg_rmse = {}
avg_mae = {}
for model_type, metrics in results.items():
    rmses = [m[0] for m in metrics]
    maes = [m[1] for m in metrics]
    avg_rmse[model_type] = np.mean(rmses)
    avg_mae[model_type] = np.mean(maes)

models = list(avg_rmse.keys())
x = np.arange(len(models))
width = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x - width/2, [avg_rmse[m] for m in models], width, label='Average RMSE')
plt.bar(x + width/2, [avg_mae[m] for m in models], width, label='Average MAE')
plt.xlabel("Model Type")
plt.ylabel("Error")
plt.title("Figure 3: Average RMSE and MAE Across Models")
plt.xticks(x, models)
plt.legend()
plt.tight_layout()
plt.show()

# Figure 4: Ensemble Prediction Plot (Actual vs. Ensemble Predicted Mid-Price)
if "Ensemble" in final_preds:
    plt.figure(figsize=(14, 8))
    plt.plot(final_actual[:1000], label="Actual", linewidth=2)
    plt.plot(final_preds["Ensemble"][:1000], label="Predicted Ensemble", alpha=0.75)
    plt.title("Figure 4: Ensemble Prediction Plot (Actual vs. Ensemble Predicted Mid-Price)")
    plt.xlabel("Samples")
    plt.ylabel("Mid-Price")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
